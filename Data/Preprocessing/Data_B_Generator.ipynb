```python
import pandas as pd
import yfinance as yf
import datetime as dt
import time
import requests
import io
```

```python
start = dt.datetime(2014,1,1)
end = dt.datetime(2022,3,1)
```

```python
sgx = pd.read_csv("SGXData_mainboard_active.csv")
```

```python
# Getting the Trading Code only and turn it into a list of string
Symbols = sgx['Trading Code'].tolist()

def add_SI(x):
  new_lst = []
  for ele in x:
    ele = ele + '.SI'
    new_lst.append(ele)
  return new_lst

Symbols = add_SI(Symbols)
Symbols.insert(0, "^STI")
Symbols
```

```python
# create empty dataframe
stock_final = pd.DataFrame()
# iterate over each symbol
for i in Symbols:  
    
    # print the symbol which is being downloaded
    print( str(Symbols.index(i)) + str(' : ') + i, sep=',', end=',', flush=True)  
    
    try:
        # download the stock price 
        stock = []
        stock = yf.download(i,start=start, end=end, interval='1mo', progress=False)
        
        # append the individual stock prices 
        if len(stock) == 0:
            None
        else:
            stock['Name']=i
            stock_final = stock_final.append(stock,sort=False)
    except Exception:
        None
```

```python
df = stock_final
df.reset_index(inplace=True)
df = df.filter(['Name', 'Date', 'Close'])
df.head()
```

```python
# Filter just jan 1
# df['Date'] = pd.to_datetime(df['Date']) -- if column is not datetime yet
new_df = df[df['Date'].dt.month == 1]
new_df
```

```python
new_df['Year'] = new_df['Date'].map(lambda x : x.year)
new_df
```

```python
# DONE?
final_df = new_df.filter(['Name', 'Year', 'Close'])
final_df
```

```python
df2 = final_df.groupby(['Name','Year']).sum()
df3 = df2.unstack(level=-1).add_prefix('p_')

output_df = df3['p_Close'].reset_index()

output_df
```

```python
output_df.to_csv("Data_B.csv")
```
