```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

```python
sgx = pd.read_csv("SGX.csv")
```

```python
def thousands_millions_negative_cleaner(data_list):
    data_list_clean = []
    
    for d in data_list:
        if d == "—" or d == "−":
            data_list_clean.append(0)
            continue
        
        multiplier = 1
        #Remove M/K sign and multiply later
        if d[-1] == "M":
            multiplier = 1000000
            d = d[:-1]
        elif d[-1] == "K":
            multiplier = 1000
            d = d[:-1]
        
        #Remove - (i.e negative sign) before value
        if d[0] == "−" or d[0] == "—":
            total = 0 - float(d[1:]) * multiplier
        else:
            total = float(d) * multiplier
            
        data_list_clean.append(total)
    
    return data_list_clean
```

```python
import time
```

```python
driver = webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver.exe")
```

```python
final = pd.DataFrame()

stocks_notfound = []

error_stocks = []

for coy in sgx['Trading Code']:
    url = f"https://www.tradingview.com/symbols/SGX-{coy}/financials-income-statement/"
    url2 = f"https://www.tradingview.com/symbols/SGX-{coy}/financials-statistics-and-ratios/"
    try:
        driver.get(url)
    except:
        stocks_notfound.append(coy)
        continue
    try:
        time.sleep(2)
        #Years
        years = driver.find_element_by_class_name("container-W9rWl7P3")
        years = years.text.split()[2:]

        #EBIT
        ebit = driver.find_elements_by_xpath('//div[@class="container-jKD0Exn- withChange-jKD0Exn-"]')
        ebit = ebit[5].text.split()[1::2]
        ebit = [e.replace("\u202a", "").replace("\u202c", "") for e in ebit]
        ebit_clean = thousands_millions_negative_cleaner(ebit)

        #Next Page
        driver.get(url2)
        time.sleep(2)
        #EV
        table = driver.find_elements_by_xpath('//div[@class="container-jKD0Exn-"]')
        ev = table[4].text.split()[2:]
        ev = [e.replace("\u202a", "").replace("\u202c", "") for e in ev]
        ev_clean = thousands_millions_negative_cleaner(ev)

        #ROC
        roc = table[7].text.split()[5:]
        roc = [r.replace("\u202a", "").replace("\u202c", "") for r in roc]
        roc_clean = thousands_millions_negative_cleaner(roc)

        #DataFrame
        new_df = pd.DataFrame()
        new_df['Company'] = ''
        new_df['Year'] = years
        new_df['ROC'] = roc_clean
        new_df['EBIT'] = ebit_clean
        new_df['EV'] = ev_clean
        new_df['Company'] = coy

    except:
        error_stocks.append(coy)
        
    #Append to final dataframe
    if len(final.columns) == 0:
        final = new_df.copy()
    else:
        final = final.append(new_df)
```
